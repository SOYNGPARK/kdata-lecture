* requests의 get방식 통해 html 가져온 후, bs 사용하기

```python
import requests

def download_rq(url, agent="python bot", num_retries=2) :
    headers = {"user-agent" : agent} 
    resp = requests.request('get', url, headers = headers)
    
    if 500 <= resp.status_code <=600 and num_retries != 0:
        print(resp.status_code, resp.reason)
        return download(url, num_retries = num_retries-1)

    return resp
```

```python
agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.99 Safari/537.36"
link = "https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%ED%95%9C%EA%B5%AD"
```

```python
html_rq = download_rq(link, agent = agent)
```

```python
html_rq.text
```

```python
doc_rq = BeautifulSoup(html_rq.text, "lxml")
```

```python
doc_rq.find('a') # 하나만
```

```python
aList_rq = doc_rq.find_all('a') # 모두
```

```python
aList_rq
```

* urllib request 통해 html 가져온 후, bs 사용하기

```python
from urllib.request import Request
from urllib.request import urlopen

def download(url, agent="python bot", num_retries=2) :
    headers = {"user-agent" : agent} 
    req = Request(url, headers = headers) 
    
    try :
        resp = urlopen(req)
        
    except HTTPError as e:
        resp = None
        
        print(e.code, e.reason, e.headers)
        
        if 500 <= e.code <=600 and num_retries != 0:
            return download(url, num_retries = num_retries-1)
        
    return resp
```

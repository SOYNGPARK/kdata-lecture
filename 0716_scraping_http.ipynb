1) urllib.request.Request

```python
from urllib.request import Request
```

```python
# 500대 에러(서버 에러)가 나면 num_retries 만큼 반복해서 urlopen을 시도하는 함수

def download(url, agent="python bot", num_retries=2) :
    headers = {"user-agent" : agent} # header에 user-agent 가 없어서 만들어줌
    req = Request(url, headers = headers) # 입력된 헤더를 가지고 요청함
    
    try :
        resp = urlopen(req)
        
    except HTTPError as e:
        resp = None
        
        print(e.code, e.reason, e.headers)
        
        if 500 <= e.code <=600 and num_retries != 0:
            return download(url, num_retries = num_retries-1)
        
    return resp
```

2) requests.request - get 방식

```python
import requests
```

```python
# requests.request - get

def download_rq(url, agent="python bot", num_retries=2) :
    headers = {"user-agent" : agent}
    resp = requests.request('get', url, headers = headers) # urlopen 과정이 필요없음
    
    if 500 <= resp.status_code <=600 and num_retries != 0:
        print(resp.status_code, resp.reason)
        return download(url, num_retries = num_retries-1)

    return resp
```

```python
html_rq = download_rq("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%ED%95%9C%EA%B5%AD", agent = agent)
```

3) requests.request - post 방식

```python
import json
```

```python
params = {
    "where" : "nexearch",
    "sm" : "top_hty",
    "fbm" : 1,
    "ie" : "utf8",
    "query" :"%ED%95%9C%EA%B5%AD"
}
```

```python
jsonParams = json.dumps(params)
```

```python
jsonParams
```

```python
# post 방식 - json으로 파라미터 넘김
def download_json(url, agent="python bot", num_retries=2) :
    headers = {"user-agent" : agent} 
    resp = requests.request('post', url, headers = headers, json=jsonParams) 
    
    if 500 <= resp.status_code <=600 and num_retries != 0:
        print(resp.status_code, resp.reason)
        return download(url, num_retries = num_retries-1)

    return resp
```

```python
dj = download_json("http://httpbin.org/post")  
# post 방식 데이터 잘 받았는지 확인하는 주소
```

```python
dj.text # \ -> escape 문자 잘 받았다는 뜻
```

```python
# get 방식
def download_get(url, agent="python bot", num_retries=2) :
    headers = {"user-agent" : agent} 
    resp = requests.get(url, headers = headers) 
    
    if 500 <= resp.status_code <=600 and num_retries != 0:
        print(resp.status_code, resp.reason)
        return download(url, num_retries = num_retries-1)

    return resp
```

5) requests.request - get 방식3

```python
# get 방식 - url의 파라미터를 get의 알규먼트로 넣어주는 경우
def download_get_params(url, agent="python bot", num_retries=2) :
    headers = {"user-agent" : agent} 
    resp = requests.get(url, headers = headers, params=params) 
    
    if 500 <= resp.status_code <=600 and num_retries != 0:
        print(resp.status_code, resp.reason)
        return download(url, num_retries = num_retries-1)

    return resp
```

```python
dgp = download_get_params("https://search.naver.com/search.naver", agent=agent) # 본주소만 적으면 된다
```

```python
url = "http://openapi.airkorea.or.kr/openapi/services/rest/ArpltnInforInqireSvc/getMsrstnAcctoRltmMesureDnsty"

params = {"serviceKey" : "u%2BuyMDJmWoL7ghFJKwN79h%2FRHI3LbLLRFex3b3RwtSiG106Cw8q36%2F2Y3l3AZ3OLpkF1Q%2BRalM7NZkglL2deBA%3D%3D",
          "numOfRows" : 10,
          "pageSize" : 10,
          "pageNo" : 1,
          "stationName" : None,
          "dataTerm" : "DAILY",
          "ver" : "1.3"}
```

```python
# servicekey byte -> string 변환
params["serviceKey"] = urllib.parse.unquote(params["serviceKey"])
```

```python
params["stationName"] = "종로구"
```

```python
params
```

```python
paramUrl = urllib.parse.urlencode(params)
```

```python
paramUrl
```

2) URL 형 변환 (str -> byte)

```python
paramBytes = paramUrl.encode('utf-8')
```

```python
paramBytes
```

3) Request & respond

```python
# 나 이 url에 있는 데이터 갖고 싶어
req = urllib.request.Request(url, data=paramBytes)
```

```python
# 내가 받아올게
res = urllib.request.urlopen(req)
```

4) 결과 읽기

```python
result = res.read()
```

```python
type(result)
```

```python
result
```

json으로 받아올때는 받아온 bytes 타입의 데이터를 string으로 변환해야 했으나, xml은 그럴필요없음

* xml etree

```python
xml = etree.fromstring(result)
```

```python
type(xml)
```

```python
# 출력
etree.dump(xml)
```
